```python
import pandas as pd
import sqlite3
```

# Bathc handle csv files

```python
dfs = [pd.read_csv(f'output{i}.csv') for i in range(8)]
df = pd.concat(dfs)
df['原文片名']=df['原文片名'].fillna('unknown')
df = df.reset_index(drop=True)
```

```python
# Manual Error fix
df.loc[df['中文片名'] == '伊人', '類型'] = 'Short'
df.loc[df['中文片名'] == '一路逆風', '類型'] = 'No Genres'
df.loc[df['中文片名'] == '八美圖', '類型'] = 'No Genres'
df.loc[df['中文片名'] == '一起去天堂', '類型'] = 'Short, Drama, Family'
df.loc[df['中文片名'] == '黑暗暖光', '類型'] = 'Short, Drama'
df.loc[df['中文片名'] == '離巢', '類型'] = 'Short, Drama'
df.loc[df['中文片名'] == '浮陽', '類型'] = 'Short, Drama, Family'
df.loc[df['中文片名'] == 'RPG', '類型'] = 'No Genres'
df.loc[df['中文片名'] == '工寮', '類型'] = 'Documentary'
df.loc[df['中文片名'] == '何以為家', '類型'] = 'Documentary, Short'
df.loc[df['中文片名'] == 'Puppy Love', '類型'] = 'No Genres'
df.loc[df['中文片名'] == '多幾咧？', '類型'] = 'Animation, Short'
df.loc[df['中文片名'] == '一步一步的愛', '類型'] = 'No Genres'
df.loc[df['中文片名'] == '回家的理由', '類型'] = 'Documentary'
df.loc[df['中文片名'] == '藍色獨白', '類型'] = 'Animation, Short'
df.loc[df['中文片名'] == '阿查依蘭的呼喚', '類型'] = 'Documentary'
df.loc[df['中文片名'] == '降河洄游', '類型'] = 'Short, Drama'
df.loc[df['中文片名'] == '削瘦的靈魂', '類型'] = 'Documentary'
df.loc[df['中文片名'] == '願未央', '類型'] = 'Documentary'
df.loc[df['中文片名'] == '邪門', '類型'] = 'Horror, Mystery'
```

```python
df['IMDB ID'] = 'unknown'
```

# 讀取其他資料來源
---
Kaggle 下載電影資料集 * 3 (已經個別清理完)

```python
Movies_clean_df = pd.read_csv('Movies_clean_data.csv')
```

# 合併、年分排序

```python
Total_df = pd.concat([df, Movies_clean_df])
Total_df = Total_df.sort_values('年度')
Total_df = Total_df.reset_index(drop=True)
```

```python
Total_df
```

# Save to DataBase

```python
# 不帶index 轉DB
conn = sqlite3.connect('Movies.db')
Total_df.to_sql('Movies_latest', con=conn, if_exists='replace', index=False)
conn.close()
```
